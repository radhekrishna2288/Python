```python
'''
3. What are the different ways to define and call a function in Python?

ans) Standard Function Definition - This is the most common way to define a function, using the def keyword, 
followed by the function name, parentheses for parameters, and a colon. The function body is indented below.
'''
def greet(name):
    """This function greets the person passed in as a parameter."""
    print(f"Hello, {name}!")

greet("Alice")

'''
Lambda Functions (Anonymous Functions)
These are small, single-expression functions defined using the lambda keyword. They are often used for short
operations and can be defined inline.
'''

add = lambda x, y: x + y
print(add(5, 3))  # Calling the lambda function

'''
Functions with Default Arguments
You can provide default values for function parameters. If a value is not provided when calling the function,
the default value will be used.
'''

def greet(name, message="Hello"):
    print(message, name)

greet("Bob")  # Calling with only one argument

'''
Functions with Variable Number of Arguments

*args is used to pass a variable number of non-keyword arguments.
**kwargs is used to pass a variable number of keyword arguments.
'''

def add_all(*numbers):
    return sum(numbers)

print(add_all(1, 2, 3, 4))  # Calling with multiple arguments

def greet(**words):
    for key, value in words.items():
        print(f"{key}: {value}")

greet(name="John", greeting="Hello")
```

```python
'''
6. Explain the concept of generators in Python and how they are defined.

ans) Regular functions calculates the values and returns in one go which takes a lot of memory and time if there 
is a lot of computation as all the computations will be done first and then the result will be returned in one go.
Here comes the concept of generators which helps to generate the result one by one instead of one go.

A generator in Python is a function that returns an iterator. It produces a sequence of values one at a time using 
the yield keyword, instead of computing and storing all values at once like a regular function with return.
'''
#Example

def sqr_num(n):
    for i in range(n+1):
        yield i**2
gen = sqr_num(3)

next(gen)
```

```python
'''
7. What are the advantages of using generators over regular functions?

ans) Regular functions calculates the values and returns in one go which takes a lot of memory and time if there 
is a lot of computation as all the computations will be done first and then the result will be returned in one go.
Here comes the concept of generators which helps to generate the result one by one instead of one go taking up less
memory space and time.
'''

# Example:

def fib (n):
    a = 0
    b = 1
    for i in range (n):
        yield a
        a, b = b, a+b
f = fib(5)
next(f)
```

```python
'''
9. Explain the purpose and usage of the 'map()" function in Python.

ans) The map() function in Python applies a given function to each item of an iterable (like a list or tuple) 
and returns an iterator that yields the results. This provides a concise way to perform operations on sequences
of data without writing explicit loops.
'''
# Example

l = [1,2,3,4,5]
def add(x):
    return x+10 # adding 10 to each of the number in list
list(map(add, l))
```

```python
'''
10. What is the difference between 'map()', 'reduce()' and 'filter()' function in Python?

ans) The map(), filter(), and reduce() functions are built-in higher-order functions in Python that operate on 
iterables (like lists, tuples, etc.). They each serve distinct purposes:

map():
Applies a given function to each item in an iterable and returns an iterator that yields the results. It 
transforms each element individually.

filter():
Creates a new iterator that filters the elements from an iterable based on a function that returns True or False.
It selects elements that meet a condition.

reduce():
Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a 
single value. It aggregates elements.
'''

#Example

numbers = [1, 2, 3, 4, 5]

# map(): squares each number
list(map(lambda x: x**2, numbers))
```

```python
# filter(): keeps only even numbers
list(filter(lambda x: x % 2 == 0, numbers))
```

```python
from functools import reduce

# reduce(): calculates the product of all numbers
reduce(lambda x, y: x * y, numbers)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def check_prime_numbers():
    prime_numbers = []
    for num in range(2, 201):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            prime_numbers.append(num)
    return prime_numbers

primes = check_prime_numbers()
print("Prime numbers between 1 and 200:", primes)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci(n):
    l = [0,1] 
    for i in range(2,n):
        l.append(l[-1]+l[-2])
    return l

fibo = fibonacci(6)
print("Fibonacci Series: ",fibo)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")

if __name__ == "__main__":
    
    with open("my_file.txt", "w") as f:
        f.write("This is line 1.\n")
        f.write("This is line 2.\n")
        f.write("This is line 3.\n")
        
    for line in read_file_lines("my_file.txt"):
        print(line.strip()) # Remove trailing newline character
```
